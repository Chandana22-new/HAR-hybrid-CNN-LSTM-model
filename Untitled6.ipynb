CNN-LSTM Model

```python
# --- 1. Correct and Necessary Imports ---
import numpy as np
from numpy import dstack
from pandas import read_csv

# All Keras/TensorFlow imports must be from tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import TimeDistributed
from tensorflow.keras.layers import Conv1D      # Corrected import path
from tensorflow.keras.layers import MaxPooling1D # Corrected import path
from tensorflow.keras.utils import to_categorical

# --- 2. Data Loading and Preprocessing Functions ---

def file_load(filepath):
    """Loads a single data file."""
    # Use 'latin-1' encoding for robustness with this dataset's structure
    df = read_csv(filepath, header=None, delim_whitespace=True, encoding='latin-1')
    return df.values

def train_test_append(filenames, append_before=''):
    """Loads and stacks multiple files into a single 3D numpy array."""
    datalist = list()
    for name in filenames:
        data = file_load(append_before + name)
        datalist.append(data)

    # Stacks the files along the third dimension (features)
    datalist = dstack(datalist)
    return datalist

def inertial_signals_load(group, append_before=''):
    """Loads all inertial signals (X) and activity labels (y) for a group."""
    filepath = append_before + group + '/Inertial Signals/'

    # 9 files total: total_acc (x,y,z), body_acc (x,y,z), body_gyro (x,y,z)
    filenames = list()
    filenames += ['total_acc_x_'+group+'.txt', 'total_acc_y_'+group+'.txt', 'total_acc_z_'+group+'.txt']
    filenames += ['body_acc_x_'+group+'.txt', 'body_acc_y_'+group+'.txt', 'body_acc_z_'+group+'.txt']
    filenames += ['body_gyro_x_'+group+'.txt', 'body_gyro_y_'+group+'.txt', 'body_gyro_z_'+group+'.txt']

    # Load and stack the features (X)
    X = train_test_append(filenames, filepath)

    # Load the labels (y)
    y = file_load(append_before + group + '/y_'+group+'.txt')
    return X, y

def load_dataset(append_before=''):
    """Main function to load and preprocess the entire dataset."""
    print("Loading Training Data...")
    trainX, trainy = inertial_signals_load('train', append_before + 'UCI HAR Dataset/')
    print("Loading Testing Data...")
    testX, testy = inertial_signals_load('test', append_before + 'UCI HAR Dataset/')

    # Convert labels from 1-6 to 0-5 (required for one-hot encoding)
    trainy = trainy - 1
    testy = testy - 1

    # One-hot encode the labels
    trainy = to_categorical(trainy)
    testy = to_categorical(testy)

    print("Data shapes after loading and one-hot encoding:")
    print(f"trainX: {trainX.shape}, trainy: {trainy.shape}")
    print(f"testX: {testX.shape}, testy: {testy.shape}")

    return trainX, trainy, testX, testy

# --- 3. Execute Loading and Reshaping ---

trainX, trainy, testX, testy = load_dataset()

# Model Hyperparameters derived from data
verbose, epochs, batch_size = 0, 400, 150
n_timesteps = trainX.shape[1]  # Should be 128
n_features = trainX.shape[2]   # Should be 9
n_outputs = trainy.shape[1]    # Should be 6

# Parameters for CNN-LSTM segmentation
n_steps = 4
n_length = 32

# Reshape the data for CNN-LSTM model: (Samples, 128, 9) -> (Samples, n_steps, n_length, n_features)
trainX = trainX.reshape((trainX.shape[0], n_steps, n_length, n_features))
testX = testX.reshape((testX.shape[0], n_steps, n_length, n_features))

print("\nFinal shapes for CNN-LSTM input:")
print(f"trainX shape: {trainX.shape}")
print(f"testX shape: {testX.shape}")

# --- 4. Corrected CNN-LSTM Model Definition and Compilation ---

# The variables n_length, n_features, and n_outputs are now correctly defined.
model = Sequential()

# CNN Layers applied via TimeDistributed
model.add(TimeDistributed(Conv1D(filters=128, kernel_size=4, activation='relu'), input_shape=(None, n_length, n_features)))
model.add(TimeDistributed(MaxPooling1D(pool_size=2)))
model.add(TimeDistributed(Conv1D(filters=128, kernel_size=4, activation='relu')))
model.add(TimeDistributed(Dropout(0.5)))
model.add(TimeDistributed(MaxPooling1D(pool_size=2)))
model.add(TimeDistributed(Flatten()))

# LSTM Layer
model.add(LSTM(100))
model.add(Dropout(0.5))

# Dense Output Layers
model.add(Dense(100, activation='relu'))
model.add(Dense(n_outputs, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

print("\nModel Summary:")
model.summary()

# --- Ready for Training ---
# To train the model, you would run:
# history = model.fit(trainX, trainy, epochs=epochs, batch_size=batch_size, validation_data=(testX, testy), verbose=1)
```

```python
# --- Model Definition, Training, and Evaluation ---

model = Sequential()
model.add(TimeDistributed(Conv1D(filters=128, kernel_size=4, activation='relu'), input_shape=(None,n_length,n_features)))
model.add(TimeDistributed(MaxPooling1D(pool_size=2)))
model.add(TimeDistributed(Conv1D(filters=128, kernel_size=4, activation='relu')))
model.add(TimeDistributed(Dropout(0.5)))
model.add(TimeDistributed(MaxPooling1D(pool_size=2)))
model.add(TimeDistributed(Flatten()))
model.add(LSTM(100))
model.add(Dropout(0.5))
model.add(Dense(100, activation='relu'))
model.add(Dense(n_outputs, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Training and Evaluation
history=model.fit(trainX, trainy, validation_data=(testX, testy), epochs=epochs, batch_size=batch_size, verbose=verbose)
loss , accuracy = model.evaluate(testX, testy, batch_size=batch_size, verbose=0)
score=accuracy*100

print('\n--- Model Results ---')
print('Accuracy = {:.2f}%'.format(score))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import metrics # Import metrics for confusion_matrix

# Get predictions and true labels
ypred_train = model.predict(trainX)
max_ypred_train = np.argmax(ypred_train, axis=1)

LABELS = ['WALKING','WALKING_UPSTAIRS','WALKING_DOWNSTAIRS','SITTING','STANDING','LAYING']

ypred_test = model.predict(testX)
max_ypred_test = np.argmax(ypred_test, axis=1)
max_ytest = np.argmax(testy, axis=1)

# Compute and plot the Confusion Matrix
matrix = metrics.confusion_matrix(max_ytest, max_ypred_test)

plt.figure(figsize=(6, 4))
sns.heatmap(matrix, cmap="YlGnBu", xticklabels=LABELS, yticklabels=LABELS, annot=True, fmt='d')
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
